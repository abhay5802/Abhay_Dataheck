```python
import pandas as pd
vaccine = pd.read_csv("training_set_features.csv")
labels = pd.read_csv("training_set_labels.csv")
test = pd.read_csv("test_set_features.csv")
vaccine.head()
```

```python
vaccine.set_index('respondent_id',inplace = True)
labels.set_index('respondent_id',inplace = True)
test.set_index('respondent_id',inplace = True)
dataset= pd.concat([vaccine,labels],axis=1)
dataset.head()
```

```python
X = dataset.drop(['xyz_vaccine','seasonal_vaccine'],axis=1)
X.head()
```

```python
y1 = dataset['xyz_vaccine']
y1.head()
```

```python
y2 = dataset['seasonal_vaccine']
y2.head()
```

```python
num_col = X.select_dtypes(include=['int64','float64']).columns
num_test = test.select_dtypes(include=['int64','float64']).columns
num_test
```

```python
cols_to_fill = [ 'xyz_concern', 'xyz_knowledge',
       'behavioral_antiviral_meds', 'behavioral_avoidance',
       'behavioral_face_mask', 'behavioral_wash_hands',
       'behavioral_large_gatherings', 'behavioral_outside_home',
       'behavioral_touch_face', 'doctor_recc_xyz', 'doctor_recc_seasonal',
       'chronic_med_condition', 'child_under_6_months', 'health_worker',
       'health_insurance', 'opinion_xyz_vacc_effective', 'opinion_xyz_risk',
       'opinion_xyz_sick_from_vacc', 'opinion_seas_vacc_effective',
       'opinion_seas_risk', 'opinion_seas_sick_from_vacc', 'household_adults',
       'household_children']
test_cols_to_fill = [ 'xyz_concern', 'xyz_knowledge',
       'behavioral_antiviral_meds', 'behavioral_avoidance',
       'behavioral_face_mask', 'behavioral_wash_hands',
       'behavioral_large_gatherings', 'behavioral_outside_home',
       'behavioral_touch_face', 'doctor_recc_xyz', 'doctor_recc_seasonal',
       'chronic_med_condition', 'child_under_6_months', 'health_worker',
       'health_insurance', 'opinion_xyz_vacc_effective', 'opinion_xyz_risk',
       'opinion_xyz_sick_from_vacc', 'opinion_seas_vacc_effective',
       'opinion_seas_risk', 'opinion_seas_sick_from_vacc', 'household_adults',
       'household_children']
```

```python
X[cols_to_fill]=X[cols_to_fill].fillna(X[cols_to_fill].median())
test[test_cols_to_fill]=test[test_cols_to_fill].fillna(test[test_cols_to_fill].median())
```

```python
X.isnull().sum()
```

```python
X.shape
```

```python
X.dropna(inplace=False).shape
```

```python
cat_col = X.select_dtypes(include=['object']).columns
test_cat_col = test.select_dtypes(include=['object']).columns
```

```python
for col in cat_col:
    mode_value=X[col].mode()[0]
    X[col]=X[col].fillna(mode_value)
for col in test_cat_col:
    mode_value=test[col].mode()[0]
    test[col]=test[col].fillna(mode_value)
```

```python
X = pd.get_dummies(X, columns=cat_col)
test = pd.get_dummies(test, columns=test_cat_col)
```

```python
X.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X[num_col]=scaler.fit_transform(X[num_col])
test[num_test]=scaler.fit_transform(test[num_test])
X.head()
```

```python
from sklearn.linear_model import LogisticRegression
model1= LogisticRegression(max_iter=1000)
model2 = LogisticRegression(max_iter=1000)
model1.fit(X,y1)
model2.fit(X,y2)
pred_probab1 = model1.predict_proba(test)
pred_probab2 = model2.predict_proba(test)
```

```python
part1 = pd.DataFrame(pred_probab1[:,1])
part2 = pd.DataFrame(pred_probab2[:,1])
part3 = pd.DataFrame(test.index)
result = pd.concat([part3,part1,part2],axis=1)
result.head()
```

```python
result = result.set_index('respondent_id')
```
